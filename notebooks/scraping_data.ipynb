```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
%matplotlib inline
import urllib.request
from bs4 import BeautifulSoup
import re



user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'

url = "https://www.worldometers.info/coronavirus/"
headers={'User-Agent':user_agent,} 

request=urllib.request.Request(url,None,headers) 
response = urllib.request.urlopen(request)
data = response.read() # The data we need
soup = BeautifulSoup(data, 'lxml')
type(soup)

text = soup.get_text()
print(soup.title.text)

```

```python
table = soup.find(name='table', attrs={'id':'main_table_countries_today'})
list_rows = []

rows = table.find_all('tr')
header = table.find_all('th')
str_cells = str(header)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
list_rows.append(cleantext)
#------------------------------------------------------

for row in rows: 
    data = row.find_all (name= 'td')
    str_cells2 = str(data)
    cleantext2 = BeautifulSoup(str_cells2, "lxml").get_text()
    list_rows.append(cleantext2)
    
for i in list_rows : 
    print (i)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(' ', expand=True)
df1.head(20)

df1[0] = df1[0].str.strip('[')
df1.head(20)

df2 = df1.rename(columns=df1.iloc[0])
df2.head(20)
```

```python
df3 = df2.iloc[10:]
df3.head(5)
```

```python
fig = go.Figure(go.Bar(
            x=df3['Country,Other,'][0:8],
            y=df3['TotalCases,'][0:8],
            orientation='v'))
fig.update_layout(
    title='total Cases In Each Country',
    template='plotly_white',
     xaxis_title="total Cases",
    yaxis_title="Countries",
)
fig.show()
```
